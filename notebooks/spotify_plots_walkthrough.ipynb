```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
spotify_df = pd.read_csv("../data/Cleaned_Spotify_2024_Global_Streaming_Data.csv")  
spotify_df.head()
```

Bar Chart

```python
#group by genre and get average streams
average_streams = spotify_df.groupby('Genre')['Total Streams (Millions)'].mean().sort_values(ascending =False)
average_streams
```

```python
plt.bar(average_streams.index, average_streams.values)
plt.show()
```

```python
plt.bar(average_streams.index, average_streams.values)
plt.xlabel('Genre')
plt.ylabel('Average Total Streams (Millions)')  
plt.title('Average Total Streams by Genre ')

plt.show()
```

```python
plt.figure(figsize=(10,6))

plt.bar(average_streams.index, average_streams.values)
plt.xlabel('Genre')
plt.ylabel('Average Total Streams (Millions)')  
plt.title('Average Total Streams by Genre ')
plt.xticks(rotation=45, ha='right' )
plt.tight_layout()

plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.bar(average_streams.index, average_streams.values)

ax = plt.gca()
ax.spines['top'].set_visible(False) 
ax.spines['right'].set_visible(False)

plt.xlabel('Genre')
plt.ylabel('Average Total Streams (Millions)')  
plt.title('Average Total Streams by Genre ')
plt.xticks(rotation=45, ha='right' )

plt.tight_layout()
plt.show()
```

```python
average_artist_streams = spotify_df.groupby('Artist')['Total Streams (Millions)'].mean().sort_values(ascending=False)
average_artist_streams
```

```python
plt.figure(figsize=(10,6))
colors = ["#6f5f9b"] * len(average_artist_streams)

max_index = average_artist_streams.idxmax()
highlight_position = average_artist_streams.index.get_loc(max_index)
colors[highlight_position] = "#ee07cf"

plt.bar(average_artist_streams.index, average_artist_streams.values, color = colors, width =0.8)
ax = plt.gca()
ax.spines['top'].set_visible(False) 
ax.spines['right'].set_visible(False)

plt.xlabel('Artist')
plt.ylabel('Average Total Streams (Millions)')  
plt.title('Average Total Streams by Artist ')
plt.xticks(rotation=90, ha='right' )

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,6))

special_artist = "Ariana Grande"
colors = ["#6f5f9b" if artist == special_artist else "gray" for artist in average_artist_streams.index]



plt.bar(average_artist_streams.index, average_artist_streams.values, color = colors, width =0.8)
ax = plt.gca()
ax.spines['top'].set_visible(False) 
ax.spines['right'].set_visible(False)

plt.xlabel('Artist')
plt.ylabel('Average Total Streams (Millions)')  
plt.title('Ariana Grande Ranks 2nd in Total Average Streams ')
plt.xticks(rotation=90, ha='right' )

plt.tight_layout()
plt.show()
```

Scatter Plot


```python
spotify_df.columns
spotify_df.info()

```

```python
plt.figure(figsize=(10,6))

plt.scatter(
    spotify_df['Streams Last 30 Days (Millions)'], 
    spotify_df['Total Streams (Millions)'],
    color = '#6f5f9b',
    alpha = 0.6,
    s = 10)
plt.xlabel('Streams This Month')
plt.ylabel('Total Streams')
plt.title('No Apparent Correlation Between Monthly Listens and Total Streams')

ax = plt.gca()
ax.spines['top'].set_visible(False) 
ax.spines['right'].set_visible(False)




plt.show()
```

```python
plt.figure(figsize=(10,6))


featured_genre = "Pop"
colors = ["#31e750" if genre == featured_genre else 'gray' for genre in spotify_df['Genre']]

plt.scatter(
    spotify_df['Streams Last 30 Days (Millions)'], 
    spotify_df['Total Streams (Millions)'],
    color = colors,
    alpha = 0.6,
    s = 10)


plt.xlabel('Streams This Month')
plt.ylabel('Total Streams')


ax = plt.gca()
ax.spines['top'].set_visible(False) 
ax.spines['right'].set_visible(False)
ax.set_title('Do Pop Songs Achieve Higher Monthly Listens or Total Streams?', fontsize = 30, pad =35)

for spine in ax.spines.values():
    spine.set_linewidth(.025)
    spine.set_alpha(0.5)
    

plt.tight_layout()
plt.show()
```

Line Chart

```python
yearly_average = spotify_df.groupby('Release Year')['Total Streams (Millions)'].mean().round(2)
yearly_average
```

```python
plt.figure(figsize=(10,6))
plt.plot(
    yearly_average.index, 
    yearly_average.values,
    marker="+",
    linestyle = "--",
    color = "#52eb24",
    linewidth = 0.5,)

plt.xlabel('Release Year')
plt.ylabel('Average Total Streams (Millions)')      
plt.title('Did Listeners Stream Less in 2022? ', fontsize = 18, pad =20)

ax = plt.gca()
ax.spines['top'].set_visible(False)     
ax.spines['right'].set_visible(False)

ax.axvline(
    2022, 
    color="#52eb24", 
    linewidth=0.5,
    alpha=0.7,)

ax.text(
    2022 + 0.05,
    2225,
    'Dip In Total Streams',
    color="#52eb24",
    fontsize = 8)
        


plt.ylim(0, 3500)

plt.show()  
```

Pie Chart

```python
category = 'Platform Type'
values = 'Total Streams (Millions)'

platform_totals = spotify_df.groupby(category)[values].sum()
platform_totals
```

```python
plt.figure(figsize=(8,8))
plt.pie(
    platform_totals,
    labels = platform_totals.index,
    startangle=130,
    colors = ['#ff9999','#66b3ff'],
    autopct='%1.1f%%',)

plt.title('Near Even Split Between Premium and Free Users')

plt.legend(
    labels = platform_totals.index,
    loc = 'lower right',
    title = 'Platform Type'
)

plt.tight_layout()
plt.show
```

```python
fig, ax = plt.subplots(figsize=(8,8))


wedges, texts, autotexts = ax.pie(
    platform_totals,
    labels = platform_totals.index,
    startangle=110,
    colors = ['#ff9999','#66b3ff'],
    autopct='%1.1f%%',)

ax.set_title(
    'Near Even Split Between Premium and Free Users',
    fontsize = 40,
    pad = 15,
    loc = 'center')

for autotext in autotexts:
    autotext.set_color('white')
    autotext.set_fontsize(16)
    autotext.set_weight('bold')

ax.legend(
    wedges,
    platform_totals.index,
    loc = 'lower center', 
    bbox_to_anchor = (0.5, 0.93),    
    ncol = 2,   
)

plt.tight_layout()
plt.show
```

Histogram

```python
plt.figure(figsize=(10,6))
plt.hist(
    spotify_df['Avg Stream Duration (Min)'],
    bins = 20,
    color = "#5f9b62",
    edgecolor = 'white',
)

plt.xlabel('Average Stream Duration (Minutes)')
plt.ylabel('Number of SongsTracks')     
plt.title('Distribution of Average Stream Duration', fontsize = 16, pad = 20)

ax = plt.gca()
for spine in ax.spines.values():
    spine.set_linewidth(0.5)
    spine.set_alpha(0.5)   
ax.spines['top'].set_visible(False) 
ax.spines['right'].set_visible(False) 

plt.show
```
